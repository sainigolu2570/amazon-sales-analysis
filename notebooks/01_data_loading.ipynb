```python
# ============================================
# CELL 1: Import all required libraries
# ============================================
# pandas - for working with data tables
# numpy - for mathematical operations
# matplotlib & seaborn - for creating charts
# os - for working with file paths

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# This line makes our charts look beautiful
sns.set_theme(style="whitegrid")

# This makes charts show inside notebook
%matplotlib inline

print("✅ All libraries imported successfully!")
```

```python
# ============================================
# CELL 2: Load the Amazon dataset
# ============================================
# We use pandas read_csv() to load CSV files
# '../data/Amazon.csv' means:
#   .. = go up one folder (from notebooks to project root)
#   data/ = enter data folder
#   Amazon.csv = our file

df = pd.read_csv('../data/Amazon.csv', encoding='latin1')

# encoding='latin1' handles special characters in the data
# Without this, you might get UnicodeDecodeError

print("✅ Dataset loaded successfully!")
print(f"📊 Total Rows: {df.shape[0]}")
print(f"📋 Total Columns: {df.shape[1]}")
```

```python
# ============================================
# CELL 3: See what our data looks like
# ============================================
# .head() shows first 5 rows
# This is always the FIRST thing a Data Analyst does

print("🔍 FIRST 5 ROWS OF DATA:")
print("="*60)
df.head()
```

```python
# ============================================
# CELL 4: Understand our columns
# ============================================
# .info() is like an X-ray of your dataset
# It shows:
#   - Column names
#   - Data types (text, number, date etc.)
#   - How many non-null (non-empty) values exist

print("📋 DATASET INFORMATION:")
print("="*60)
df.info()
```

```python
# ============================================
# CELL 5: Find missing/empty values
# ============================================
# Missing values = empty cells in your data
# .isnull().sum() counts empty cells per column

print("❓ MISSING VALUES PER COLUMN:")
print("="*60)
missing = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100

# Create a summary table
missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing Percentage': missing_percent.round(2)
})

# Only show columns that actually have missing values
missing_df = missing_df[missing_df['Missing Count'] > 0]
print(missing_df.sort_values('Missing Percentage', ascending=False))
```

```python
# ============================================
# CELL 6: Basic statistics of numeric columns
# ============================================
# .describe() gives us:
#   count, mean, min, max, std for every number column
# This helps us spot outliers and understand ranges

print("📈 BASIC STATISTICS:")
print("="*60)
df.describe()
```
