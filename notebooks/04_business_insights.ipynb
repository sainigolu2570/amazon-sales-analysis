```python
# ============================================
# AMAZON SALES — BUSINESS INSIGHTS REPORT
# ============================================
import pandas as pd
import numpy as np

df = pd.read_csv('../data/Amazon_cleaned.csv')
df['date'] = pd.to_datetime(df['date'])

print("="*60)
print("   AMAZON SALES ANALYSIS — BUSINESS INSIGHTS REPORT")
print("="*60)

# ── INSIGHT 1: Overall Business Health ──
print("\n📊 INSIGHT 1: OVERALL BUSINESS HEALTH")
print("-"*40)
print(f"  Total Orders    : {len(df):,}")
print(f"  Total Revenue   : ₹{df['revenue'].sum():,.2f}")
print(f"  Avg Order Value : ₹{df['amount'].mean():,.2f}")
print(f"  Total Units Sold: {df['qty'].sum():,}")

# ── INSIGHT 2: Best Performing Month ──
print("\n📅 INSIGHT 2: BEST PERFORMING MONTH")
print("-"*40)
monthly = df.groupby('month_name')['revenue'].sum()
best_month = monthly.idxmax()
worst_month = monthly.idxmin()
print(f"  🏆 Best Month  : {best_month} "
      f"(₹{monthly[best_month]:,.2f})")
print(f"  📉 Worst Month : {worst_month} "
      f"(₹{monthly[worst_month]:,.2f})")
print(f"  💡 Insight: Focus marketing budget "
      f"heavily in {best_month}")

# ── INSIGHT 3: Top Category ──
print("\n🏷️  INSIGHT 3: CATEGORY PERFORMANCE")
print("-"*40)
cat_rev = df.groupby('category')['revenue'].sum().sort_values(
    ascending=False)
print(f"  🥇 Top Category    : {cat_rev.index[0]} "
      f"(₹{cat_rev.iloc[0]:,.2f})")
print(f"  🥈 2nd Category    : {cat_rev.index[1]} "
      f"(₹{cat_rev.iloc[1]:,.2f})")
print(f"  🥉 3rd Category    : {cat_rev.index[2]} "
      f"(₹{cat_rev.iloc[2]:,.2f})")
top3_share = (cat_rev.iloc[:3].sum()/cat_rev.sum()*100)
print(f"  💡 Top 3 categories contribute "
      f"{top3_share:.1f}% of total revenue")

# ── INSIGHT 4: Geographic Analysis ──
print("\n🗺️  INSIGHT 4: GEOGRAPHIC INSIGHTS")
print("-"*40)
state_rev = df.groupby('ship_state')['revenue'].sum(
    ).sort_values(ascending=False)
print(f"  🏆 Top State    : {state_rev.index[0]} "
      f"(₹{state_rev.iloc[0]:,.2f})")
print(f"  📦 Total States : {df['ship_state'].nunique()}")
top3_states_share = (state_rev.iloc[:3].sum()/
                     state_rev.sum()*100)
print(f"  💡 Top 3 states contribute "
      f"{top3_states_share:.1f}% of revenue")
print(f"  💡 Recommendation: Open fulfillment "
      f"centers near {state_rev.index[0]}")

# ── INSIGHT 5: Order Status Health ──
print("\n📦 INSIGHT 5: ORDER STATUS HEALTH")
print("-"*40)
status = df['status'].value_counts(normalize=True)*100
for s, pct in status.head(4).items():
    print(f"  {s:25s}: {pct:.1f}%")
if 'Cancelled' in status.index:
    print(f"\n  ⚠️  Cancellation Rate: "
          f"{status.get('Cancelled', 0):.1f}%")
    if status.get('Cancelled', 0) > 10:
        print("  🚨 HIGH cancellation rate! "
              "Investigate root cause.")
    else:
        print("  ✅ Cancellation rate is acceptable.")

# ── INSIGHT 6: Quarterly Trends ──
print("\n📈 INSIGHT 6: QUARTERLY PERFORMANCE")
print("-"*40)
quarterly = df.groupby('quarter')['revenue'].sum()
best_q = quarterly.idxmax()
print(f"  🏆 Best Quarter  : Q{best_q} "
      f"(₹{quarterly[best_q]:,.2f})")
for q, rev in quarterly.items():
    bar = "█" * int(rev/quarterly.max()*20)
    print(f"  Q{q}: {bar} ₹{rev:,.0f}")

# ── INSIGHT 7: Order Size Analysis ──
print("\n💰 INSIGHT 7: ORDER SIZE BREAKDOWN")
print("-"*40)
size_pct = df['order_size'].value_counts(normalize=True)*100
for size, pct in size_pct.items():
    rev = df[df['order_size']==size]['revenue'].sum()
    print(f"  {size:15s}: {pct:.1f}% of orders | "
          f"₹{rev:,.0f} revenue")

print("\n" + "="*60)
print("   END OF BUSINESS INSIGHTS REPORT")
print("="*60)
```
